```python
pd.set_option("display.max_columns",200)
```

```python
cs_data=pd.read_csv("Credit_Score_Clean.csv")
```

```python

le = LabelEncoder()
cs_data["Occupation"]= le.fit_transform(cs_data["Occupation"])
```

```python
Oe=OrdinalEncoder()
a=["Credit_Mix","Payment_of_Min_Amount","Payment_Behaviour","Credit_Score"]
for colums in a:

    cs_data[colums] = Oe.fit_transform(np.array(cs_data[colums]).reshape(-1,1))
      
```

```python
x=cs_data.drop("Credit_Score",axis=1)
y=cs_data["Credit_Score"]
```

```python
Scaler=StandardScaler()
arr=["Age","Annual_Income","Changed_Credit_Limit","Outstanding_Debt","Credit_Utilization_Ratio","Total_EMI_per_month","Amount_invested_monthly","Monthly_Balance"]
for col in arr:
    cs_data[col] = Scaler.fit_transform(np.array(cs_data[col]).reshape(-1,1))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
RFC=RandomForestClassifier()
```

```python
RFC.fit(x_train,y_train)
```

```python
y_pred = RFC.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)*100
```
